```python
### QUESTION 4 ###

import numpy as np
import math

### Solving Methods ###

# Bisection method
def bisection(f, a, b, tol, maxiter):
    
    xsol = np.zeros(maxiter); # vector containing the approx. sols
    ek = 0
    ekp = 0 
    
    for i in np.arange(0, maxiter):
        c = (a+b)/2
        
        if (f(a)*f(c) <= 0):
            xsol[i] = c;
            b = c;
        elif (f(a)*f(c) > 0):
            xsol[i] = c;
            a = c;
                    
        # check tolerance stopping condition
        if (i > 0):
            if (abs(xsol[i] - xsol[i-1]) < tol):
                xsol = xsol[0:i];
                break;
    
    return xsol


# Secant method
def secant(f, a, b, tol, maxiter):
    
    xsol = np.zeros(maxiter); # vector containing the approx. sols
    
    for i in np.arange(0, maxiter):
        if (i == 0): # initial guess
            xsol[i] = b - (f(b) * (b - a)/(f(b) - f(a)));
        elif (i == 1):
            xsol[i] = xsol[i-1] - f(xsol[i-1]) * (xsol[i-1] - b)/(f(xsol[i-1]) - f(b));
        else:
            xsol[i] = xsol[i-1] - f(xsol[i-1]) * (xsol[i-1] - xsol[i-2])/(f(xsol[i-1]) - f(xsol[i-2]));
            
        # check tolerance stopping condition
        if (i > 0):
            if (abs(xsol[i] - xsol[i-1]) < tol):
                xsol = xsol[0:i];
                break;
    
    return xsol


# Newton method
def newton(f, fprime, x0, tol, maxiter):
    
    xsol = np.zeros(maxiter); # vector containing the approx. sols
    
    for i in np.arange(0, maxiter):
        if (i == 0):
            xsol[i] = x0 - f(x0)/fprime(x0);
        else:
            xsol[i] = xsol[i-1] - f(xsol[i-1])/fprime(xsol[i-1]);
        
        # check tolerance stopping condition
        if (i > 0):
            if (abs(xsol[i] - xsol[i-1]) < tol):
                xsol = xsol[0:i];
                break;
    
    return xsol


# Fixed Point method
def fixedpoint(g, x0, tol, maxiter):
    
    xsol = np.zeros(maxiter); # vector containing the approx. sols
    
    for i in np.arange(0, maxiter):
        if (i == 0):
            xsol[i] = g(x0);
        else:
            xsol[i] = g(xsol[i-1]);
        
        # check tolerance stopping condition
        if (i > 0):
            if (abs(xsol[i] - xsol[i-1]) < tol):
                xsol = xsol[0:i];
                break;
    
    return xsol
             
    
### Setup Tests ###

def f1(x):
    return (x**2) - 5

def f1prime(x):
    return (2*x)

# function for fixed point: g(x) = (x + 5) / (x + 1) 
def g1(x):
    return (x + 5) / (x + 1);

# function to print error table
def errtable(xsol):
    for i in np.arange(0, len(xsol)):
        ek = abs(xsol[i] - math.sqrt(5))
        if (i == 0):
            print("Iteration %d: \t x_k: %0.7e \t e_k: %0.7e \t ratio1: - \t\t\t ratio2: - \t ratio3: - \n" % 
                  (i+1, xsol[i], ek));
        else:
            rat1 = ek / (abs(xsol[i-1] - math.sqrt(5)));
            rat2 = ek / ((abs(xsol[i-1] - math.sqrt(5))) ** ((1 + math.sqrt(5)) / 2));
            rat3 = ek / ((abs(xsol[i-1] - math.sqrt(5))) ** 2);
            print("Iteration %d: \t x_k: %0.7e \t e_k: %0.7e \t ratio1: %0.7e \t\t\t ratio2: %0.7e \t ratio3: %0.7e \n" %
                  (i+1, xsol[i], ek, rat1, rat2, rat3));
        

# approx. sol vectors from each method
bis = bisection(f1, 1, 4, 1e-10, 100)
sec = secant(f1, 1, 4, 1e-10, 100)
new = newton(f1, f1prime, 1, 1e-10, 100)
fix = fixedpoint(g1, 1, 1e-10, 100)

# print error tables for each method
print("Bisection Method\n")
errtable(bis)
print("Secant Method\n")
errtable(sec)
print("Netwon Method\n")
errtable(new)
print("Fixed Point Method\n")
errtable(fix)
            
```

```python
### QUESTION 5 ###

from scipy.stats import norm
import numpy as np
import math

#adapted from Octave's financial toolkit
def blsprice(Price, Strike, Rate, Time, Volatility):
    sigma_sqrtT = Volatility * np.sqrt (Time)

    d1 = 1 / sigma_sqrtT * (np.log(Price / Strike) + (Rate + Volatility**2 / 2) * Time)
    d2 = d1 - sigma_sqrtT

    phi1 = norm.cdf(d1)
    phi2 = norm.cdf(d2)
    disc = np.exp (-Rate * Time)
    F    = Price * np.exp ((Rate) * Time)

    Call = disc * (F * phi1 - Strike * phi2)
    Put  = disc * (Strike * (1 - phi2) + F * (phi1 - 1))
    return Call, Put

#adapted from Octave's financial toolkit 
def blsdelta(Price, Strike, Rate, Time, Volatility):
    d1 = 1 / (Volatility * np.sqrt(Time)) * (np.log (Price / Strike) + (Rate + Volatility**2 / 2) * Time)

    phi = norm.cdf(d1)

    CallDelta = phi
    PutDelta = phi - 1
    return CallDelta, PutDelta


# part a)

# fixedm will solve the warrant pricing problem using fixed point
# iteration scheme and provide a table of step k vs. Wk and Rk
# S0 - share price
# K - strike
# T - expiry time
# r - interest rate
# sigma - volatility
# M - number of warrants outstanding
# N - number of original shares outstanding
def fixedm(S0, K, T, r, sigma, M, N):
    
    Wk = 0; # current warrant price, starts at Wk = 0 
    C = blsprice(S0 + (M/N)*Wk, K, r, T, sigma)[0] # call value
    Wk1 = (N / (N+M)) * C; # warrant price at step k+1 (starts at step 0)
    Rk = Wk - Wk1 # residual at step k (starts at step 0)
    i = 0; # step number, starts at i = 0
    print("Fixed Point Iteration:\n")
    print("Step %d: \t Wk: %0.7f \t Rk: %0.7f \n" % 
                      (i, Wk, Rk));

    while (abs(Wk1 - Wk) >= 10**(-8)):
        i = i + 1;
        Wk = Wk1
        C = blsprice(S0 + (M/N)*Wk, K, r, T, sigma)[0];
        Wk1 = (N / (N+M)) * C;
        Rk = Wk - Wk1;
        print("Step %d: \t Wk: %0.7f \t Rk: %0.7e \n" % 
                      (i, Wk, Rk));
    return Wk1


Wfixprice = fixedm(100, 100, 2, 0.04, 0.3, 7 * (10**6), 30 * (10**6))
print("\n")


# part b)

# newtonm will solve the warrant pricing problem using newton method
# and provide a table of step k vs. Wk and Rk
# S0 - share price
# K - strike
# T - expiry time
# r - interest rate
# sigma - volatility
# M - number of warrants outstanding
# N - number of original shares outstanding
def newtonm(S0, K, T, r, sigma, M, N):
    
    Wk = 0; # current warrant price, starts at Wk = 0 
    C = blsprice(S0 + (M/N)*Wk, K, r, T, sigma)[0] # call value
    Cdelta = blsdelta(S0 + (M/N)*Wk, K, r, T, sigma)[0] # call value of blsprice derivative
    
    f1 = Wk - (N / (N+M))*C # function f(W) for newton method
    f1prime = 1 - (M / (N+M))*Cdelta # derivative of function f(W) for newton method
    
    Wk1 = Wk - f1/f1prime; # warrant price at step k+1 (starts at step 0)
    Rk = Wk - Wk1 # Rk - residual at step k (starts at step 0)
    i = 0; # i - step number, starts at i = 0
    print("Newton Method:\n")
    print("Step %d: \t Wk: %0.7f \t Rk: %0.7f \n" % 
                      (i, Wk, Rk));

    while (abs(Wk1 - Wk) >= 10**(-8)):
        i = i + 1;
        Wk = Wk1
        C = blsprice(S0 + (M/N)*Wk, K, r, T, sigma)[0];
        Cdelta = blsdelta(S0 + (M/N)*Wk, K, r, T, sigma)[0];
        
        f1 = Wk - (N / (N+M))*C
        f1prime = 1 - (M / (N+M))*Cdelta
        
        Wk1 = Wk - f1/f1prime;
        Rk = Wk - Wk1;
        print("Step %d: \t Wk: %0.7f \t Rk: %0.7e \n" % 
                      (i, Wk, Rk));
    return Wk1


Wnewprice = newtonm(100, 100, 2, 0.04, 0.3, 7 * (10**6), 30 * (10**6))



```
